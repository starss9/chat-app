```python
from google.colab import files
import pandas as pd
from sklearn.ensemble import IsolationForest
import joblib

uploaded= files.upload()

filename = list(uploaded.keys())[0]

# Load your CSV data
df = pd.read_csv('box-optimiser-code - Sheet1.csv')

# Example: Assuming you have columns named 'volume' and 'weight'
X = df[['Length (cm)', 'Width (cm)']]

# Train the Isolation Forest
model = IsolationForest(contamination=0.05, random_state=42)
model.fit(X)

# Save the trained model
joblib.dump(model, 'anomaly_model.pkl')

# Optional: Add anomaly predictions to your data for review
df['anomaly'] = model.predict(X)  # -1 is anomaly, 1 is normal
print(df.head())
```
